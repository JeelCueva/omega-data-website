````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# ConfiguraciÃ³n de visualizaciÃ³n
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (14, 8)
plt.rcParams['font.size'] = 10

# ==============================================================================
# 1. CARGA Y LIMPIEZA DE DATOS
# ==============================================================================

print("="*80)
print("ANÃLISIS ECONOMÃ‰TRICO DE COSTOS DE PARCELAS - CHULLAY")
print("="*80)

# Cargar datos
file_path = '/content/COSTOS_DE_20_PARCELAS-CHULLAY-PERIODO_1,_2_Y_3(1).xlsx'
df_resumen = pd.read_excel(file_path, sheet_name='RESUMEN')

# Limpiar datos
df = df_resumen.dropna(subset=['Periodo', 'Parcelas']).copy()
print(f"\nâœ“ Datos cargados: {len(df)} observaciones de {df['Parcelas'].nunique()} parcelas")

# Verificar estructura
print(f"\nColumnas disponibles:")
for i, col in enumerate(df.columns, 1):
    print(f"  {i}. {col}")

# ==============================================================================
# 2. ANÃLISIS EXPLORATORIO DE DATOS (EDA)
# ==============================================================================

print("\n" + "="*80)
print("ESTADÃSTICAS DESCRIPTIVAS")
print("="*80)

# EstadÃ­sticas por periodo
print("\nğŸ“Š RESUMEN POR PERIODO:")
resumen_periodo = df.groupby('Periodo').agg({
    'Area total m2': ['mean', 'median', 'std'],
    'Mano de obra': ['mean', 'sum'],
    'Semillas': ['mean', 'sum'],
    'Fertilizantes': ['mean', 'sum'],
    'Pesticidas': ['mean', 'sum'],
    'TOTAL CD': ['mean', 'sum'],
    'TOTAL CI': ['mean', 'sum'],
    'TOTAL': ['mean', 'sum', 'std']
}).round(2)
print(resumen_periodo)

# EstadÃ­sticas generales
print("\nğŸ“ˆ ESTADÃSTICAS GENERALES DE COSTOS:")
costos_cols = ['Mano de obra', 'Semillas', 'Fertilizantes', 'Pesticidas',
               'Equipos y herramientas', 'Agua', 'Otros insumos',
               'TOTAL CD', 'TOTAL CI', 'TOTAL']

stats_df = df[costos_cols].describe().T
stats_df['CV'] = (stats_df['std'] / stats_df['mean'] * 100).round(2)  # Coeficiente de variaciÃ³n
print(stats_df)

# ==============================================================================
# 3. ANÃLISIS DE COSTOS POR CATEGORÃA
# ==============================================================================

print("\n" + "="*80)
print("ANÃLISIS DE ESTRUCTURA DE COSTOS")
print("="*80)

# ComposiciÃ³n promedio de costos directos
costos_directos = ['Mano de obra', 'Semillas', 'Fertilizantes', 'Pesticidas',
                   'Equipos y herramientas', 'Agua', 'Otros insumos']
composicion_cd = df[costos_directos].mean()
print("\nğŸ’° COMPOSICIÃ“N DE COSTOS DIRECTOS (Promedio en S/):")
for item, valor in composicion_cd.items():
    pct = (valor / composicion_cd.sum() * 100)
    print(f"  {item:.<30} S/ {valor:>8.2f} ({pct:>5.1f}%)")

# Costos por hectÃ¡rea
df['Costo_por_ha'] = df['TOTAL'] / (df['Area total m2'] / 10000)
df['CD_por_ha'] = df['TOTAL CD'] / (df['Area total m2'] / 10000)
df['CI_por_ha'] = df['TOTAL CI'] / (df['Area total m2'] / 10000)

print("\nğŸ“ COSTOS POR HECTÃREA:")
print(f"  Costo Total Promedio: S/ {df['Costo_por_ha'].mean():.2f}/ha")
print(f"  Costo Directo Promedio: S/ {df['CD_por_ha'].mean():.2f}/ha")
print(f"  Costo Indirecto Promedio: S/ {df['CI_por_ha'].mean():.2f}/ha")

# ==============================================================================
# 4. ANÃLISIS ECONOMÃ‰TRICO - MODELOS DE REGRESIÃ“N
# ==============================================================================

print("\n" + "="*80)
print("MODELOS ECONOMÃ‰TRICOS")
print("="*80)

# Preparar datos para regresiÃ³n
df['log_area'] = np.log(df['Area total m2'] + 1)
df['log_total'] = np.log(df['TOTAL'] + 1)
df['periodo_dummy'] = df['Periodo'].astype(str)

# Crear dummies para periodo
periodo_dummies = pd.get_dummies(df['Periodo'], prefix='Periodo', drop_first=True)
df_reg = pd.concat([df, periodo_dummies], axis=1)

# ==============================================================================
# MODELO 1: Determinantes del Costo Total
# ==============================================================================

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

print("\n" + "-"*80)
print("MODELO 1: DETERMINANTES DEL COSTO TOTAL")
print("-"*80)

# Variables independientes
X_vars = ['Area total m2', 'Mano de obra', 'Semillas', 'Fertilizantes',
          'Pesticidas', 'Equipos y herramientas']
X = df[X_vars].fillna(0)
y = df['TOTAL']

# Ajustar modelo
modelo1 = LinearRegression()
modelo1.fit(X, y)
y_pred1 = modelo1.predict(X)

# Resultados
r2_1 = r2_score(y, y_pred1)
rmse_1 = np.sqrt(mean_squared_error(y, y_pred1))

print(f"\nğŸ“Š RESULTADOS DEL MODELO:")
print(f"  RÂ² = {r2_1:.4f}")
print(f"  RMSE = S/ {rmse_1:.2f}")
print(f"  RÂ² Ajustado = {1 - (1-r2_1)*(len(y)-1)/(len(y)-X.shape[1]-1):.4f}")

print(f"\nğŸ”¢ COEFICIENTES:")
print(f"  {'Variable':<30} {'Coeficiente':>15} {'Impacto':>15}")
print("-"*65)
for var, coef in zip(X_vars, modelo1.coef_):
    print(f"  {var:<30} {coef:>15.4f} {coef:>15.4f}")
print(f"  {'Intercepto':<30} {modelo1.intercept_:>15.4f}")

# ==============================================================================
# MODELO 2: Efecto del Ãrea en los Costos (Log-Log)
# ==============================================================================

print("\n" + "-"*80)
print("MODELO 2: ELASTICIDAD DEL COSTO RESPECTO AL ÃREA")
print("-"*80)

X2 = df[['log_area']].fillna(0)
y2 = df['log_total']

modelo2 = LinearRegression()
modelo2.fit(X2, y2)
y_pred2 = modelo2.predict(X2)

r2_2 = r2_score(y2, y_pred2)
elasticidad = modelo2.coef_[0]

print(f"\nğŸ“Š RESULTADOS DEL MODELO LOG-LOG:")
print(f"  RÂ² = {r2_2:.4f}")
print(f"  Elasticidad = {elasticidad:.4f}")
print(f"\nğŸ’¡ INTERPRETACIÃ“N:")
print(f"  Un incremento del 1% en el Ã¡rea genera un incremento del {elasticidad:.2f}% en el costo total")
if elasticidad < 1:
    print(f"  âœ“ EconomÃ­as de escala: Los costos crecen menos que proporcional al Ã¡rea")
elif elasticidad > 1:
    print(f"  âœ— DeseconomÃ­as de escala: Los costos crecen mÃ¡s que proporcional al Ã¡rea")
else:
    print(f"  = Costos proporcionales al Ã¡rea")

# ==============================================================================
# MODELO 3: AnÃ¡lisis de Panel - Efectos Fijos por Periodo
# ==============================================================================

print("\n" + "-"*80)
print("MODELO 3: PANEL DATA - EFECTOS FIJOS POR PERIODO")
print("-"*80)

# Preparar variables para panel
panel_vars = ['Area total m2', 'Mano de obra', 'Semillas']
if 'Periodo_2.0' in df_reg.columns and 'Periodo_3.0' in df_reg.columns:
    panel_vars.extend(['Periodo_2.0', 'Periodo_3.0'])

X3 = df_reg[panel_vars].fillna(0)
y3 = df_reg['TOTAL']

modelo3 = LinearRegression()
modelo3.fit(X3, y3)
y_pred3 = modelo3.predict(X3)

r2_3 = r2_score(y3, y_pred3)

print(f"\nğŸ“Š RESULTADOS DEL MODELO PANEL:")
print(f"  RÂ² = {r2_3:.4f}")

print(f"\nğŸ”¢ COEFICIENTES:")
for var, coef in zip(panel_vars, modelo3.coef_):
    if 'Periodo' in var:
        print(f"  {var:<30} {coef:>15.2f} (Efecto temporal)")
    else:
        print(f"  {var:<30} {coef:>15.4f}")

# ==============================================================================
# 5. ANÃLISIS DE VARIANZA (ANOVA)
# ==============================================================================

print("\n" + "="*80)
print("ANÃLISIS DE VARIANZA (ANOVA)")
print("="*80)

# ANOVA para diferencias entre periodos
from scipy.stats import f_oneway

periodo1 = df[df['Periodo'] == 1]['TOTAL'].dropna()
periodo2 = df[df['Periodo'] == 2]['TOTAL'].dropna()
periodo3 = df[df['Periodo'] == 3]['TOTAL'].dropna()

f_stat, p_value = f_oneway(periodo1, periodo2, periodo3)

print(f"\nğŸ“Š PRUEBA ANOVA - Diferencias entre Periodos:")
print(f"  F-estadÃ­stico: {f_stat:.4f}")
print(f"  P-valor: {p_value:.4f}")

if p_value < 0.05:
    print(f"  âœ“ Existen diferencias significativas entre los costos de los periodos (Î±=0.05)")
else:
    print(f"  âœ— No hay diferencias significativas entre los costos de los periodos (Î±=0.05)")

# ==============================================================================
# 6. CORRELACIONES
# ==============================================================================

print("\n" + "="*80)
print("MATRIZ DE CORRELACIONES")
print("="*80)

corr_vars = ['Area total m2', 'Mano de obra', 'Semillas', 'Fertilizantes',
             'Pesticidas', 'TOTAL CD', 'TOTAL']
corr_matrix = df[corr_vars].corr()

print("\n", corr_matrix.round(3))

# ==============================================================================
# 7. IDENTIFICACIÃ“N DE OUTLIERS
# ==============================================================================

print("\n" + "="*80)
print("ANÃLISIS DE VALORES ATÃPICOS (OUTLIERS)")
print("="*80)

def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower) | (data[column] > upper)]
    return outliers, lower, upper

outliers_total, lower_total, upper_total = find_outliers_iqr(df, 'TOTAL')

print(f"\nğŸ“ OUTLIERS EN COSTO TOTAL:")
print(f"  LÃ­mite inferior: S/ {lower_total:.2f}")
print(f"  LÃ­mite superior: S/ {upper_total:.2f}")
print(f"  NÃºmero de outliers: {len(outliers_total)}")

if len(outliers_total) > 0:
    print(f"\n  Parcelas con valores atÃ­picos:")
    for idx, row in outliers_total.iterrows():
        print(f"    - {row['Parcelas']}: S/ {row['TOTAL']:.2f} (Periodo {int(row['Periodo'])})")

# ==============================================================================
# 8. RESUMEN EJECUTIVO DE MODELOS
# ==============================================================================

print("\n" + "="*80)
print("RESUMEN COMPARATIVO DE MODELOS")
print("="*80)

resumen_modelos = pd.DataFrame({
    'Modelo': [
        'Modelo 1: Determinantes del Costo Total',
        'Modelo 2: Elasticidad (Log-Log)',
        'Modelo 3: Panel Data - Efectos Fijos'
    ],
    'RÂ²': [r2_1, r2_2, r2_3],
    'Variables': [len(X_vars), 1, len(panel_vars)],
    'Observaciones': [len(X), len(X2), len(X3)]
})

print("\n", resumen_modelos.to_string(index=False))

print("\n" + "="*80)
print("âœ“ ANÃLISIS ECONOMÃ‰TRICO COMPLETADO")
print("="*80)

# Guardar resultados para usar en visualizaciones
results = {
    'df': df,
    'df_reg': df_reg,
    'modelo1': modelo1,
    'modelo2': modelo2,
    'modelo3': modelo3,
    'r2_scores': [r2_1, r2_2, r2_3],
    'X_vars': X_vars,
    'panel_vars': panel_vars,
    'corr_matrix': corr_matrix,
    'resumen_periodo': resumen_periodo,
    'composicion_cd': composicion_cd
}

print("\nğŸ’¾ Datos procesados guardados para visualizaciÃ³n...")
````

````python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Visualizaciones del AnÃ¡lisis EconomÃ©trico de Parcelas - Chullay
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# ConfiguraciÃ³n
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Set2")
plt.rcParams['font.size'] = 10
plt.rcParams['figure.dpi'] = 100

# Cargar datos
file_path = '/content/COSTOS_DE_20_PARCELAS-CHULLAY-PERIODO_1,_2_Y_3(1).xlsx'
df = pd.read_excel(file_path, sheet_name='RESUMEN')
df = df.dropna(subset=['Periodo', 'Parcelas']).copy()

# Calcular costos por hectÃ¡rea
df['Area_ha'] = df['Area total m2'] / 10000
df.loc[df['Area_ha'] > 0, 'Costo_por_ha'] = df['TOTAL'] / df['Area_ha']
df.loc[df['Area_ha'] > 0, 'CD_por_ha'] = df['TOTAL CD'] / df['Area_ha']

# ==============================================================================
# FIGURA 1: DISTRIBUCIÃ“N DE COSTOS POR PERIODO
# ==============================================================================

fig1, axes = plt.subplots(2, 2, figsize=(16, 12))
fig1.suptitle('AnÃ¡lisis de DistribuciÃ³n de Costos por Periodo', fontsize=16, fontweight='bold', y=0.995)

# 1.1 Boxplot de Costos Totales por Periodo
ax1 = axes[0, 0]
df.boxplot(column='TOTAL', by='Periodo', ax=ax1, patch_artist=True)
ax1.set_title('DistribuciÃ³n de Costos Totales por Periodo', fontsize=12, fontweight='bold')
ax1.set_xlabel('Periodo', fontsize=11)
ax1.set_ylabel('Costo Total (S/)', fontsize=11)
ax1.get_figure().suptitle('')  # Remover tÃ­tulo automÃ¡tico
for i in [1, 2, 3]:
    periodo_data = df[df['Periodo'] == i]['TOTAL']
    media = periodo_data.mean()
    ax1.axhline(media, color='red', linestyle='--', alpha=0.3, linewidth=1)

# 1.2 EvoluciÃ³n promedio de costos
ax2 = axes[0, 1]
evolucion = df.groupby('Periodo')[['Mano de obra', 'Semillas', 'Fertilizantes',
                                     'Pesticidas', 'TOTAL CD', 'TOTAL CI']].mean()
evolucion.plot(ax=ax2, marker='o', linewidth=2.5)
ax2.set_title('EvoluciÃ³n de Componentes de Costo Promedio', fontsize=12, fontweight='bold')
ax2.set_xlabel('Periodo', fontsize=11)
ax2.set_ylabel('Costo Promedio (S/)', fontsize=11)
ax2.legend(title='Componente', fontsize=9, title_fontsize=10, loc='upper left')
ax2.grid(True, alpha=0.3)

# 1.3 Histograma de Costos Totales
ax3 = axes[1, 0]
for periodo in [1, 2, 3]:
    data = df[df['Periodo'] == periodo]['TOTAL']
    ax3.hist(data, alpha=0.6, label=f'Periodo {int(periodo)}', bins=10, edgecolor='black')
ax3.set_title('DistribuciÃ³n de Frecuencias de Costos Totales', fontsize=12, fontweight='bold')
ax3.set_xlabel('Costo Total (S/)', fontsize=11)
ax3.set_ylabel('Frecuencia', fontsize=11)
ax3.legend(fontsize=10)
ax3.grid(True, alpha=0.3, axis='y')

# 1.4 Violin plot
ax4 = axes[1, 1]
periodo_str = df['Periodo'].astype(int).astype(str)
sns.violinplot(x=periodo_str, y='TOTAL', data=df.assign(Periodo_str=periodo_str), ax=ax4, palette='Set2')
ax4.set_title('Densidad de DistribuciÃ³n de Costos por Periodo', fontsize=12, fontweight='bold')
ax4.set_xlabel('Periodo', fontsize=11)
ax4.set_ylabel('Costo Total (S/)', fontsize=11)

plt.tight_layout()
plt.savefig('/content/fig1_distribucion_costos.png', dpi=300, bbox_inches='tight')
print("âœ“ Figura 1 guardada: DistribuciÃ³n de costos por periodo")

# ==============================================================================
# FIGURA 2: COMPOSICIÃ“N Y ESTRUCTURA DE COSTOS
# ==============================================================================

fig2, axes = plt.subplots(2, 2, figsize=(16, 12))
fig2.suptitle('AnÃ¡lisis de ComposiciÃ³n y Estructura de Costos', fontsize=16, fontweight='bold', y=0.995)

# 2.1 Pie chart - ComposiciÃ³n promedio de costos directos
ax1 = axes[0, 0]
costos_directos = ['Mano de obra', 'Semillas', 'Fertilizantes', 'Pesticidas',
                   'Equipos y herramientas', 'Agua']
valores = df[costos_directos].mean()
colores = sns.color_palette("Set3", len(valores))
wedges, texts, autotexts = ax1.pie(valores, labels=valores.index, autopct='%1.1f%%',
                                     colors=colores, startangle=90, textprops={'fontsize': 9})
ax1.set_title('ComposiciÃ³n de Costos Directos (Promedio)', fontsize=12, fontweight='bold')
for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontweight('bold')

# 2.2 Barras apiladas - Costos por periodo
ax2 = axes[0, 1]
costos_periodo = df.groupby('Periodo')[costos_directos].mean()
costos_periodo.plot(kind='bar', stacked=True, ax=ax2, width=0.7)
ax2.set_title('Costos Directos Promedio por Periodo (Apilados)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Periodo', fontsize=11)
ax2.set_ylabel('Costo (S/)', fontsize=11)
ax2.legend(title='Componente', fontsize=8, title_fontsize=9, bbox_to_anchor=(1.05, 1))
ax2.set_xticklabels([f'P{int(x)}' for x in costos_periodo.index], rotation=0)
ax2.grid(True, alpha=0.3, axis='y')

# 2.3 ComparaciÃ³n CD vs CI
ax3 = axes[1, 0]
cd_ci = df.groupby('Periodo')[['TOTAL CD', 'TOTAL CI']].mean()
x = np.arange(len(cd_ci))
width = 0.35
bars1 = ax3.bar(x - width/2, cd_ci['TOTAL CD'], width, label='Costos Directos', color='steelblue')
bars2 = ax3.bar(x + width/2, cd_ci['TOTAL CI'], width, label='Costos Indirectos', color='coral')
ax3.set_title('ComparaciÃ³n: Costos Directos vs Indirectos', fontsize=12, fontweight='bold')
ax3.set_xlabel('Periodo', fontsize=11)
ax3.set_ylabel('Costo Promedio (S/)', fontsize=11)
ax3.set_xticks(x)
ax3.set_xticklabels([f'Periodo {int(i)}' for i in cd_ci.index])
ax3.legend(fontsize=10)
ax3.grid(True, alpha=0.3, axis='y')

# Agregar valores en las barras
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax3.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.0f}', ha='center', va='bottom', fontsize=9)

# 2.4 Heatmap de costos por parcela y periodo
ax4 = axes[1, 1]
pivot_data = df.pivot_table(values='TOTAL', index='Parcelas', columns='Periodo', aggfunc='sum')
sns.heatmap(pivot_data, annot=True, fmt='.0f', cmap='YlOrRd', ax=ax4, cbar_kws={'label': 'Costo Total (S/)'})
ax4.set_title('Mapa de Calor: Costos por Parcela y Periodo', fontsize=12, fontweight='bold')
ax4.set_xlabel('Periodo', fontsize=11)
ax4.set_ylabel('Parcela', fontsize=11)

plt.tight_layout()
plt.savefig('/content/fig2_composicion_costos.png', dpi=300, bbox_inches='tight')
print("âœ“ Figura 2 guardada: ComposiciÃ³n y estructura de costos")

# ==============================================================================
# FIGURA 3: ANÃLISIS DE REGRESIÃ“N Y CORRELACIONES
# ==============================================================================

fig3, axes = plt.subplots(2, 2, figsize=(16, 12))
fig3.suptitle('AnÃ¡lisis de RegresiÃ³n y Relaciones Entre Variables', fontsize=16, fontweight='bold', y=0.995)

# 3.1 Scatter plot: Ãrea vs Costo Total
ax1 = axes[0, 0]
colores_periodo = {1: 'blue', 2: 'green', 3: 'red'}
for periodo in [1, 2, 3]:
    data_periodo = df[df['Periodo'] == periodo]
    ax1.scatter(data_periodo['Area total m2'], data_periodo['TOTAL'],
               label=f'Periodo {int(periodo)}', alpha=0.6, s=80,
               color=colores_periodo[periodo], edgecolors='black', linewidth=0.5)

# LÃ­nea de regresiÃ³n
z = np.polyfit(df['Area total m2'], df['TOTAL'], 1)
p = np.poly1d(z)
x_line = np.linspace(df['Area total m2'].min(), df['Area total m2'].max(), 100)
ax1.plot(x_line, p(x_line), "r--", alpha=0.8, linewidth=2, label=f'RegresiÃ³n lineal')

ax1.set_title('RelaciÃ³n entre Ãrea y Costo Total', fontsize=12, fontweight='bold')
ax1.set_xlabel('Ãrea Total (mÂ²)', fontsize=11)
ax1.set_ylabel('Costo Total (S/)', fontsize=11)
ax1.legend(fontsize=9)
ax1.grid(True, alpha=0.3)

# Calcular RÂ²
from sklearn.metrics import r2_score
r2 = r2_score(df['TOTAL'], p(df['Area total m2']))
ax1.text(0.05, 0.95, f'RÂ² = {r2:.4f}', transform=ax1.transAxes,
         fontsize=11, verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# 3.2 Scatter plot: Mano de obra vs Costo Total
ax2 = axes[0, 1]
for periodo in [1, 2, 3]:
    data_periodo = df[df['Periodo'] == periodo]
    ax2.scatter(data_periodo['Mano de obra'], data_periodo['TOTAL'],
               label=f'Periodo {int(periodo)}', alpha=0.6, s=80,
               color=colores_periodo[periodo], edgecolors='black', linewidth=0.5)

z2 = np.polyfit(df['Mano de obra'], df['TOTAL'], 1)
p2 = np.poly1d(z2)
x_line2 = np.linspace(df['Mano de obra'].min(), df['Mano de obra'].max(), 100)
ax2.plot(x_line2, p2(x_line2), "r--", alpha=0.8, linewidth=2)

ax2.set_title('RelaciÃ³n entre Mano de Obra y Costo Total', fontsize=12, fontweight='bold')
ax2.set_xlabel('Mano de Obra (S/)', fontsize=11)
ax2.set_ylabel('Costo Total (S/)', fontsize=11)
ax2.legend(fontsize=9)
ax2.grid(True, alpha=0.3)

r2_2 = r2_score(df['TOTAL'], p2(df['Mano de obra']))
ax2.text(0.05, 0.95, f'RÂ² = {r2_2:.4f}', transform=ax2.transAxes,
         fontsize=11, verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# 3.3 Matriz de correlaciÃ³n (heatmap)
ax3 = axes[1, 0]
corr_vars = ['Area total m2', 'Mano de obra', 'Semillas', 'Fertilizantes',
             'Pesticidas', 'TOTAL CD', 'TOTAL']
corr_matrix = df[corr_vars].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool), k=1)
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.3f', cmap='coolwarm',
            center=0, ax=ax3, square=True, linewidths=1, cbar_kws={'label': 'CorrelaciÃ³n'})
ax3.set_title('Matriz de Correlaciones', fontsize=12, fontweight='bold')

# 3.4 Residuales del modelo principal
ax4 = axes[1, 1]
from sklearn.linear_model import LinearRegression

X_vars = ['Area total m2', 'Mano de obra', 'Semillas', 'Fertilizantes',
          'Pesticidas', 'Equipos y herramientas']
X = df[X_vars].fillna(0)
y = df['TOTAL']

modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)
residuales = y - y_pred

ax4.scatter(y_pred, residuales, alpha=0.6, s=60, edgecolors='black', linewidth=0.5)
ax4.axhline(y=0, color='r', linestyle='--', linewidth=2)
ax4.set_title('GrÃ¡fico de Residuales del Modelo', fontsize=12, fontweight='bold')
ax4.set_xlabel('Valores Predichos (S/)', fontsize=11)
ax4.set_ylabel('Residuales (S/)', fontsize=11)
ax4.grid(True, alpha=0.3)

# Agregar lÃ­neas de Â±2 desviaciones estÃ¡ndar
std_res = residuales.std()
ax4.axhline(y=2*std_res, color='orange', linestyle=':', linewidth=1.5, alpha=0.7, label='Â±2Ïƒ')
ax4.axhline(y=-2*std_res, color='orange', linestyle=':', linewidth=1.5, alpha=0.7)
ax4.legend(fontsize=9)

plt.tight_layout()
plt.savefig('/content/fig3_regresion_correlaciones.png', dpi=300, bbox_inches='tight')
print("âœ“ Figura 3 guardada: AnÃ¡lisis de regresiÃ³n y correlaciones")

# ==============================================================================
# FIGURA 4: ANÃLISIS POR COMPONENTE Y EFICIENCIA
# ==============================================================================

fig4, axes = plt.subplots(2, 2, figsize=(16, 12))
fig4.suptitle('AnÃ¡lisis Detallado por Componente y Eficiencia', fontsize=16, fontweight='bold', y=0.995)

# 4.1 Barras comparativas de componentes principales
ax1 = axes[0, 0]
componentes = ['Mano de obra', 'Semillas', 'Fertilizantes', 'Pesticidas', 'Equipos y herramientas']
medias = df[componentes].mean().sort_values(ascending=False)
colores_comp = sns.color_palette("viridis", len(medias))
bars = ax1.barh(range(len(medias)), medias.values, color=colores_comp, edgecolor='black', linewidth=1.2)
ax1.set_yticks(range(len(medias)))
ax1.set_yticklabels(medias.index, fontsize=10)
ax1.set_xlabel('Costo Promedio (S/)', fontsize=11)
ax1.set_title('Ranking de Componentes de Costo', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3, axis='x')

# Agregar valores en las barras
for i, (bar, valor) in enumerate(zip(bars, medias.values)):
    ax1.text(valor + 3, bar.get_y() + bar.get_height()/2, f'S/ {valor:.2f}',
            va='center', fontsize=10, fontweight='bold')

# 4.2 Serie de tiempo de costos por parcela seleccionadas
ax2 = axes[0, 1]
parcelas_principales = df.groupby('Parcelas')['TOTAL'].sum().nlargest(5).index
for parcela in parcelas_principales:
    data_parcela = df[df['Parcelas'] == parcela].sort_values('Periodo')
    ax2.plot(data_parcela['Periodo'], data_parcela['TOTAL'], marker='o',
            label=parcela, linewidth=2.5, markersize=8)

ax2.set_title('EvoluciÃ³n de Costos - Top 5 Parcelas', fontsize=12, fontweight='bold')
ax2.set_xlabel('Periodo', fontsize=11)
ax2.set_ylabel('Costo Total (S/)', fontsize=11)
ax2.legend(title='Parcela', fontsize=9, title_fontsize=10)
ax2.grid(True, alpha=0.3)
ax2.set_xticks([1, 2, 3])

# 4.3 Boxplot comparativo de componentes
ax3 = axes[1, 0]
componentes_box = ['Mano de obra', 'Semillas', 'Fertilizantes', 'Pesticidas']
df_melted = df[componentes_box].melt(var_name='Componente', value_name='Costo')
sns.boxplot(data=df_melted, x='Componente', y='Costo', ax=ax3, palette='Set2')
ax3.set_title('DistribuciÃ³n de Costos por Componente', fontsize=12, fontweight='bold')
ax3.set_xlabel('Componente de Costo', fontsize=11)
ax3.set_ylabel('Costo (S/)', fontsize=11)
ax3.tick_params(axis='x', rotation=45)
ax3.grid(True, alpha=0.3, axis='y')

# 4.4 AnÃ¡lisis de eficiencia (Costo por mÂ²)
ax4 = axes[1, 1]
df['Costo_por_m2'] = df['TOTAL'] / df['Area total m2']
df_eficiencia = df[df['Area total m2'] > 0]  # Filtrar Ã¡reas vÃ¡lidas

periodo_str = df_eficiencia['Periodo'].astype(int).astype(str)
sns.violinplot(x=periodo_str, y='Costo_por_m2', data=df_eficiencia.assign(Periodo_str=periodo_str),
               ax=ax4, palette='pastel')
ax4.set_title('Eficiencia: Costo por mÂ² por Periodo', fontsize=12, fontweight='bold')
ax4.set_xlabel('Periodo', fontsize=11)
ax4.set_ylabel('Costo por mÂ² (S/)', fontsize=11)
ax4.grid(True, alpha=0.3, axis='y')

# Agregar medias
for i, periodo in enumerate([1, 2, 3]):
    media = df_eficiencia[df_eficiencia['Periodo'] == periodo]['Costo_por_m2'].median()
    ax4.text(i, media, f'{media:.4f}', ha='center', va='bottom', fontweight='bold', fontsize=9)

plt.tight_layout()
plt.savefig('/content/fig4_componentes_eficiencia.png', dpi=300, bbox_inches='tight')
print("âœ“ Figura 4 guardada: AnÃ¡lisis por componente y eficiencia")

# ==============================================================================
# FIGURA 5: DASHBOARD EJECUTIVO
# ==============================================================================

fig5 = plt.figure(figsize=(18, 10))
gs = fig5.add_gridspec(3, 3, hspace=0.3, wspace=0.3)
fig5.suptitle('Dashboard Ejecutivo - AnÃ¡lisis EconomÃ©trico de Parcelas', fontsize=18, fontweight='bold', y=0.98)

# 5.1 KPIs principales (texto)
ax1 = fig5.add_subplot(gs[0, :])
ax1.axis('off')

kpis_text = f"""
â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
â•‘                                    INDICADORES CLAVE DE DESEMPEÃ‘O (KPIs)                            â•‘
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£
â•‘                                                                                                      â•‘
â•‘  ğŸ“Š COSTO TOTAL PROMEDIO: S/ {df['TOTAL'].mean():.2f}    ğŸ“ˆ DESVIACIÃ“N ESTÃNDAR: S/ {df['TOTAL'].std():.2f}    ğŸ“‰ COEF. VARIACIÃ“N: {(df['TOTAL'].std()/df['TOTAL'].mean()*100):.1f}%  â•‘
â•‘                                                                                                      â•‘
â•‘  ğŸ’° COSTO DIRECTO PROMEDIO: S/ {df['TOTAL CD'].mean():.2f}  ({df['TOTAL CD'].mean()/df['TOTAL'].mean()*100:.1f}%)         â•‘
â•‘  ğŸ—ï¸  COSTO INDIRECTO PROMEDIO: S/ {df['TOTAL CI'].mean():.2f}  ({df['TOTAL CI'].mean()/df['TOTAL'].mean()*100:.1f}%)          â•‘
â•‘                                                                                                      â•‘
â•‘  ğŸ‘· MANO DE OBRA (Principal): S/ {df['Mano de obra'].mean():.2f} ({df['Mano de obra'].mean()/df['TOTAL CD'].mean()*100:.1f}% del CD)    â•‘
â•‘  ğŸŒ± SEMILLAS: S/ {df['Semillas'].mean():.2f}   ğŸ§ª FERTILIZANTES: S/ {df['Fertilizantes'].mean():.2f}   â˜ ï¸  PESTICIDAS: S/ {df['Pesticidas'].mean():.2f}         â•‘
â•‘                                                                                                      â•‘
â•‘  ğŸ“ ÃREA PROMEDIO: {df['Area total m2'].mean():.0f} mÂ²           ğŸ“¦ PARCELAS ANALIZADAS: {df['Parcelas'].nunique()}                   â•‘
â•‘  â±ï¸  PERIODOS: {int(df['Periodo'].min())} - {int(df['Periodo'].max())}                     ğŸ”¬ OBSERVACIONES TOTALES: {len(df)}                      â•‘
â•‘                                                                                                      â•‘
â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
"""
ax1.text(0.5, 0.5, kpis_text, ha='center', va='center', fontsize=11, family='monospace',
         bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.2))

# 5.2 GrÃ¡fico de tendencia general
ax2 = fig5.add_subplot(gs[1, 0])
tendencia = df.groupby('Periodo')['TOTAL'].agg(['mean', 'std', 'count'])
ax2.errorbar(tendencia.index, tendencia['mean'], yerr=tendencia['std'],
            marker='o', markersize=10, linewidth=2.5, capsize=5, capthick=2)
ax2.fill_between(tendencia.index,
                tendencia['mean'] - tendencia['std'],
                tendencia['mean'] + tendencia['std'],
                alpha=0.2)
ax2.set_title('Tendencia de Costos', fontsize=11, fontweight='bold')
ax2.set_xlabel('Periodo')
ax2.set_ylabel('Costo Promedio (S/)')
ax2.grid(True, alpha=0.3)
ax2.set_xticks([1, 2, 3])

# 5.3 Top 5 parcelas por costo
ax3 = fig5.add_subplot(gs[1, 1])
top_parcelas = df.groupby('Parcelas')['TOTAL'].sum().nlargest(5)
colors_top = sns.color_palette("Reds_r", len(top_parcelas))
bars_top = ax3.barh(range(len(top_parcelas)), top_parcelas.values, color=colors_top, edgecolor='black')
ax3.set_yticks(range(len(top_parcelas)))
ax3.set_yticklabels(top_parcelas.index)
ax3.set_xlabel('Costo Total Acumulado (S/)')
ax3.set_title('Top 5 Parcelas por Costo', fontsize=11, fontweight='bold')
for bar, val in zip(bars_top, top_parcelas.values):
    ax3.text(val, bar.get_y() + bar.get_height()/2, f'S/ {val:.0f}',
            va='center', ha='left', fontsize=9, fontweight='bold')

# 5.4 DistribuciÃ³n de costos directos
ax4 = fig5.add_subplot(gs[1, 2])
cd_comp = df[['Mano de obra', 'Semillas', 'Fertilizantes', 'Pesticidas', 'Equipos y herramientas']].mean()
explode = (0.1, 0, 0, 0, 0)  # Destacar mano de obra
colors_pie = sns.color_palette("pastel", len(cd_comp))
ax4.pie(cd_comp, labels=cd_comp.index, autopct='%1.1f%%', startangle=90,
       explode=explode, colors=colors_pie, textprops={'fontsize': 8})
ax4.set_title('ComposiciÃ³n Costos Directos', fontsize=11, fontweight='bold')

# 5.5 ComparaciÃ³n CD vs CI por periodo
ax5 = fig5.add_subplot(gs[2, 0])
cd_ci_periodo = df.groupby('Periodo')[['TOTAL CD', 'TOTAL CI']].mean()
cd_ci_periodo.plot(kind='bar', ax=ax5, color=['steelblue', 'coral'], width=0.7)
ax5.set_title('CD vs CI por Periodo', fontsize=11, fontweight='bold')
ax5.set_xlabel('Periodo')
ax5.set_ylabel('Costo Promedio (S/)')
ax5.set_xticklabels([f'P{int(x)}' for x in cd_ci_periodo.index], rotation=0)
ax5.legend(['Costos Directos', 'Costos Indirectos'], fontsize=8)
ax5.grid(True, alpha=0.3, axis='y')

# 5.6 DispersiÃ³n Ã¡rea vs costo
ax6 = fig5.add_subplot(gs[2, 1])
scatter = ax6.scatter(df['Area total m2'], df['TOTAL'],
                     c=df['Periodo'], cmap='viridis', s=100, alpha=0.6, edgecolors='black')
ax6.set_title('Ãrea vs Costo Total', fontsize=11, fontweight='bold')
ax6.set_xlabel('Ãrea (mÂ²)')
ax6.set_ylabel('Costo Total (S/)')
cbar = plt.colorbar(scatter, ax=ax6)
cbar.set_label('Periodo')
ax6.grid(True, alpha=0.3)

# 5.7 Resumen estadÃ­stico de modelos
ax7 = fig5.add_subplot(gs[2, 2])
ax7.axis('off')

# Recalcular RÂ² de los modelos
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

X_vars = ['Area total m2', 'Mano de obra', 'Semillas', 'Fertilizantes', 'Pesticidas', 'Equipos y herramientas']
X1 = df[X_vars].fillna(0)
y1 = df['TOTAL']
modelo1 = LinearRegression().fit(X1, y1)
r2_1 = r2_score(y1, modelo1.predict(X1))

modelos_text = f"""
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚ RESUMEN DE MODELOS            â”‚
â”œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¤
â”‚                               â”‚
â”‚ MODELO 1: Determinantes       â”‚
â”‚   RÂ² = {r2_1:.4f}                   â”‚
â”‚   Variables: {len(X_vars)}                 â”‚
â”‚                               â”‚
â”‚ HALLAZGOS PRINCIPALES:        â”‚
â”‚                               â”‚
â”‚ â€¢ Mano de obra es el          â”‚
â”‚   componente dominante        â”‚
â”‚   ({df['Mano de obra'].mean()/df['TOTAL CD'].mean()*100:.0f}% de costos directos)   â”‚
â”‚                               â”‚
â”‚ â€¢ EconomÃ­as de escala         â”‚
â”‚   detectadas (elasticidad     â”‚
â”‚   < 1)                        â”‚
â”‚                               â”‚
â”‚ â€¢ No hay diferencias          â”‚
â”‚   significativas entre        â”‚
â”‚   periodos (ANOVA p>0.05)     â”‚
â”‚                               â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
"""
ax7.text(0.1, 0.5, modelos_text, ha='left', va='center', fontsize=9, family='monospace',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.3))

plt.savefig('/content/fig5_dashboard_ejecutivo.png', dpi=300, bbox_inches='tight')
print("âœ“ Figura 5 guardada: Dashboard ejecutivo")

print("\n" + "="*80)
print("âœ“ TODAS LAS VISUALIZACIONES HAN SIDO GENERADAS EXITOSAMENTE")
print("="*80)
print("\nArchivos generados:")
print("  1. fig1_distribucion_costos.png")
print("  2. fig2_composicion_costos.png")
print("  3. fig3_regresion_correlaciones.png")
print("  4. fig4_componentes_eficiencia.png")
print("  5. fig5_dashboard_ejecutivo.png")
````
